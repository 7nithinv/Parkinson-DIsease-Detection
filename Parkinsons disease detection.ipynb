```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('Parkinsson.csv')
```

```python
df.head()
```

```python
#renaming the columns
df.rename(columns=({'MDVP:Fo(Hz)':'avg_fre', 'MDVP:Fhi(Hz)':'max_fre', 'MDVP:Flo(Hz)':'min_fre', 'MDVP:Jitter(%)':'var_fre1',
       'MDVP:Jitter(Abs)':'var_fre2', 'MDVP:RAP':'var_fre3', 'MDVP:PPQ':'var_fre4', 'Jitter:DDP':'var_fre5',
       'MDVP:Shimmer':'var_amp1', 'MDVP:Shimmer(dB)':'var_amp2', 'Shimmer:APQ3':'var_amp3', 'Shimmer:APQ5':'var_amp4',
       'MDVP:APQ':'var_amp5', 'Shimmer:DDA':'var_amp6'}),inplace=True)
```

```python
# the name column is of no use while building the model

df.drop(columns = 'name', axis = 1, inplace = True)
```

```python
#checking for outliers
df.skew()
```

```python
import matplotlib.pyplot as plt
```

```python
df.hist(figsize = (25,20))
plt.show()
```

```python
#seperating features and labels
x=df.loc[:,df.columns!='status'].values[:,1:]
x1=df.loc[:,df.columns!='status']
y=df.loc[:,'status'].values
y1=df.loc[:,'status']
```

```python
#removing skewness using power transformer

from sklearn.preprocessing import PowerTransformer
pt=PowerTransformer(method='yeo-johnson') 
X_power=pt.fit_transform(x1)
df_dependent=pd.DataFrame(X_power,columns=x1.columns)
```

```python
df_dependent.skew()
```

```python
df_dependent.hist(figsize = (25,20))
plt.show()
```

```python
import seaborn as sns
```

```python
#correlation matrix

correl=df_dependent.corr()
plt.figure(figsize=(20,20))
sns.heatmap(correl,annot=True,cmap='OrRd')
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
# scaling the values using minmaxscaler

scaler=MinMaxScaler((-1,1))
x_dependent = scaler.fit_transform(df_dependent)
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x_dependent, y1, test_size = 0.2)
```

```python
pip install xgboost
```

```python
#using xgboost for training

from xgboost import XGBClassifier
```

```python
model = XGBClassifier()
model.fit(xtrain,ytrain)
```

```python
predicted = model.predict(xtest)
```

```python
# using confusion matrix for the predicted and actual values

from sklearn.metrics import confusion_matrix
cm=confusion_matrix(ytest,predicted)
plt.figure(figsize=(8,6))
fg=sns.heatmap(cm,annot=True,cmap="Reds")
figure=fg.get_figure()
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title("Output Confusion Matrix")
```
